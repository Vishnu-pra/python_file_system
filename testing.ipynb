```python
from llama_cpp import Llama
import sys
```

```python
is_gpu_available = False
if is_gpu_available == "True":
    llm = Llama(model_path="./llama-2-7b-chat.ggmlv3.q4_1.bin", n_gpu_layers=128, verbose=False)
else:
    llm = Llama(model_path="./llama-2-7b-chat.ggmlv3.q4_1.bin", n_gpu_layers=0, verbose=False)
```

```python
def get_prompt(message: str, system_prompt: str) -> str:
    texts = [f'[INST] <<SYS>>\n{system_prompt}\n<</SYS>>\n']
    texts.append(f'User: {message.strip()} [/INST]\nAssistant: ')
    return ''.join(texts)
```

```python
def stream_output(question, system):
    prompt = get_prompt(question, system)
    genai = llm(prompt, max_tokens=16000, temperature=0.75, top_k=50, top_p=0.95,  
                repeat_penalty=1.1,  stop=["[/INST]", "|/", "User:", "[User]:"], echo=False, stream=True)
    print("[Llama2]: ", end="", flush=True)
    result = ""
    for w in genai:
        word = w['choices'][0]['text']
        print(word, end="", flush=True)
        result += word
    print()
    print(result)
```

```python
system_prompt = input("[Define your bot behaviour] : ")
while True:
    user_prompt = input("[User]: ")
    stream_output(user_prompt, system_prompt)
```
